```python
import pandas as pd
import pfmanager.engine as pm
import pfmanager.currency as cu
from datetime import date
```

```python
datos=pd.read_csv("Proc_Operaciones.csv",parse_dates=[0])
```

```python
datos
```

```python
datos.dtypes
```

```python
symbols=datos["TIKR"].unique()
```

```python
symbols
```

```python
symbols_list=list(symbols)
```

```python
symbols_list
```

```python
dicc = {"symbol":[], "name":[],"currency":[]}    
```

```python
for symbol in symbols_list:
    filtro=datos[datos["TIKR"]==symbol][["TIKR","Valor","Moneda"]]
    dicc["symbol"].append(symbol)
    dicc["name"].append(filtro.iloc[0].at["Valor"])
    dicc["currency"].append(filtro.iloc[0].at["Moneda"])
    
```

```python
dicc
```

------ CREAMOS PORTFOLIO, CUENTA Y ACTIVOS -----

```python
pf=pm.Portfolio("De Giro")
acc=pm.Account("Cuenta de accciones")
acc.register_portfolio(pf)
```

```python
stock_deprecated = ["TLP","MDP"]
```

```python
for i in range(len(symbols_list)-1):
    try:
        if symbols_list[i] in stock_deprecated: validate = False
        else: validate = True
        print("Importando Asset: ",dicc["name"][i]," - ",dicc["symbol"][i] , " - ",dicc["currency"][i])        
        pf.register_asset(pm.AssetEquity(dicc["name"][i],dicc["currency"][i],dicc["symbol"][i],validate=validate))
        print("Asset importado con éxito\n")
    except:
        print("No se ha podido importar el asset. Se continua con el siguiente\n")
        continue
```
