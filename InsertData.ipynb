```python
import pandas as pd
import pfmanager.engine as pm
import pfmanager.currency as cu
from datetime import date
```

```python
datos=pd.read_csv("Proc_Operaciones.csv",parse_dates=[0])
```

```python
symbols=datos["TIKR"].unique()
```

```python
symbols_list=list(symbols)
```

```python
symbols_list
```

```python
dicc = {"symbol":[], "name":[],"currency":[]}    
```

```python
for symbol in symbols_list:
    filtro=datos[datos["TIKR"]==symbol][["TIKR","Valor","Moneda"]]
    dicc["symbol"].append(symbol)
    dicc["name"].append(filtro.iloc[0].at["Valor"])
    dicc["currency"].append(filtro.iloc[0].at["Moneda"])
    
```

```python
dicc
```

------ CREAMOS PORTFOLIO, CUENTA Y ACTIVOS -----

```python
pf=pm.Portfolio("De Giro")
acc=pm.Account("Cuenta de accciones")
acc.register_portfolio(pf)
```

```python
stock_deprecated = ["TLP","MDP"]
```

```python
for i in range(len(symbols_list)-1):
    try:
        if symbols_list[i] in stock_deprecated: validate = False
        else: validate = True
        print("Importando Asset: ",dicc["name"][i]," - ",dicc["symbol"][i] , " - ",dicc["currency"][i])        
        pf.register_asset(pm.AssetEquity(dicc["name"][i],dicc["currency"][i],dicc["symbol"][i],validate=validate))
        print("Asset importado con éxito\n")
    except:
        print("No se ha podido importar el asset. Se continua con el siguiente\n")
        continue
```

```python
for registro in datos.iterrows():    
    info = registro[1]
    currency=info["Moneda"]
    tipo_oper = info["Operación"]
    unidades = info["Unidades"]
    tikr= info["TIKR"]
    fecha = info["Fecha"]
    p_unitario_asset = info["Precio unitario"]
    p_unitario_local= info["Precio unitario (€)"]
    precio= cu.Currency(p_unitario_asset,currency,p_unitario_local,"EUR")
    
    if info["Comisiones"] > 0:
        comisiones_asset=info["Comisiones"]
        comisiones_local=info["Comisiones (€)"]
        comisiones = cu.Currency(comisiones_asset,currency,comisiones_local,"EUR")
    else:
        comisiones=0        
    
    if info["Retenciones (€)"]>0:
        taxes_local=info["Retenciones (€)"]
        taxes_asset= taxes_local * (p_unitario_asset/p_unitario_local)
        retenciones = cu.Currency(taxes_asset,currency,taxes_local,"EUR")
    else:
        retenciones =0   
    
    print("\n\nTransacción ", tipo_oper, " - TIKR: ", tikr, " - Fecha: ",fecha, "\nValor: ", unidades*precio, " - Comisiones: ",comisiones," - Retenciones: ",retenciones)
    try:
        if tipo_oper == "BUY":        
            tr = pm.TransactionBuy(unidades,precio,commissions=comisiones,taxes=retenciones,date_transaction=fecha )
            pf.get_asset(symbol=tikr).register_transaction(tr)        
        elif tipo_oper == "SELL":
            tr = pm.TransactionSell(unidades,precio,commissions=comisiones,taxes=retenciones,date_transaction=fecha )
            pf.get_asset(symbol=tikr).register_transaction(tr)
        elif tipo_oper == "DIV":
            tr = pm.TransactionDividend(unidades*precio,commissions=comisiones,taxes=retenciones,date_transaction=fecha )
            pf.get_asset(symbol=tikr).register_transaction(tr)
        print("Registrada con éxito")
    except:
        print("Ha ocurrido un error y no se ha registrado")
        
```
